```python
%%time
# gbq로 bigquery에서 데이터 불러오기
# 아래처럼 데이터를 바로 불러오기에는 데이터량이 너무 많음

query ='''
select *
from `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2017`
where extract(year from pickup_datetime) = 2017
'''

df = pd.read_gbq(query = query, dialect = 'standard',project_id = "public-data-projects")
```

```python
%%time
# 필요한 데이터를 쿼리로 미리 정제
# 시간기준으로 group by 하고 운행 건 수 카운트 집계
query ='''
select datetime_trunc(pickup_datetime, hour) as datetime,
    count(*) as cnt
from `bigquery-public-data.new_york_taxi_trips.tlc_yellow_trips_2017`
where extract(year from pickup_datetime) = 2017
group by datetime
order by datetime
'''

df = pd.read_gbq(query = query, dialect = 'standard',project_id = "public-data-projects")
```

```python
# 미리보기
df.tail()
```

```python
# 월 칼럼 만들어 groupby
df['month'] = df['datetime'].dt.month
df_month = df.groupby(['month'])[['cnt']].sum()
```

```python
# barplot으로 월별 사용량 비교
df_month.plot.bar()
```

```python
# 일별로 볼 수 있도록 date칼럼 생성
df['date'] = df['datetime'].dt.date
# 3월 데이터만 groupby 해서 df_march에 할당
df_march = df[df['month'] == 3].groupby(['date'])[['cnt']].sum()
df_march.tail()
```

```python
# 3월 14일에 이용량 급감한 것 감지
plt.figure(figsize = (10,5))
plt.xticks(rotation = -20)
sns.lineplot(data = df_march)
```

```python
# 시간대별로 보기위해 hour칼럼 만들고 분석을 위한 테이블 df_month_hour에 할당
df['hour'] = df['datetime'].dt.hour
df_month_hour = df[['month','hour','cnt']]
```

```python
# 퇴근, 저녁시간, 출근, 오후 순으로 사용량이 많음
# 월별로 사용량의 차이는 있지만 패턴의 차이는 크지 않은 것으로 보임
plt.figure(figsize = (10,5))

sns.lineplot(data = df_month_hour, x = 'hour', y = 'cnt', hue = 'month',
             legend='full', palette= 'Set2')
```
